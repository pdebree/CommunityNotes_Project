# Creating Sample of Community Notes Data 

This file creates sample of community notes data to be run in debugging the community notes scoring algorithm that can be found in the https://github.com/twitter/communitynotes repo. 

This file must be in the same directory as the original data (as listed below) in order to create new samples of the data. 

```python
import pandas as pd
```

```python
# Read in full data sets (from 
batSignals = pd.read_csv("batSignals-00000.tsv", sep='\t')
notes = pd.read_csv("notes-00000.tsv", sep='\t')
noteStat = pd.read_csv("noteStatusHistory-00000.tsv", sep='\t')
rats = pd.read_csv("ratings-00000.tsv", sep='\t')
userE = pd.read_csv("userEnrollment-00000.tsv", sep='\t')
```

```python
# Create samples of data
bat_test = batSignals.sample(frac=0.01)
notes_test = notes.sample(frac=0.01)
notestat_test = noteStat.sample(frac=0.01)
rats_test = rats.sample(frac=0.01)
userE_test = userE.sample(frac=0.01)
```

```python
# Write sample datasets to file, with current `data` directory. 
bat_test.to_csv("batSignals-test.tsv", sep='\t', index=False)
notes_test.to_csv("notes-test.tsv", sep='\t', index=False)
notestat_test.to_csv("noteStatusHistory-test.tsv", sep='\t', index=False)
rats_test.to_csv("ratings-test.tsv", sep='\t', index=False)
userE_test.to_csv("userEnrollment-test.tsv", sep='\t', index=False)
```
